```python
import pandas as pd
sys_df = pd.read_csv("us-covid-19-share-people-vaccinated.csv")
unwanted_regions = ["American Samoa", "Federated States of Micronesia"]
new_df = sys_df[sys_df['Entity'].isin(unwanted_regions) == False]

national_percentage = new_df.groupby('Day').mean()

new_df.head()

# April 8th = April 7th cumulative + average over 2 weeks of daily increase
# CI with smallest and largest increment

case_df = pd.read_csv("United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv")
case_df.loc[case_df['state'] == 'ME']

```

```python
import pandas as pd, seaborn as sns, datetime
#create new DF for only US state cases (no territories) and change submission date to datetime object

all_names = set(case_df["state"])
territories = ("DC", "FSM", "GU", "NYC", "PW", "RMI", "PR", "AS", "MP", "VI")
states = all_names.difference(territories)


us_cases = case_df[case_df["state"].isin(states)]

us_cases['submission_date']= pd.to_datetime(us_cases['submission_date'])
```

```python
#separate cases by region

us_cases_by_region = us_cases.copy()

west = ['HI','AK','WA', 'OR', 'MT', 'WY', 'ID', 'CA', 'NV', 'UT', 'CO']
southwest = ['AZ', 'NM', 'TX', 'OK']
midwest = ['ND', 'SD', 'NE', 'KS', 'MO','WI', 'MN', 'IA', 'IL', 'IN','OH', 'MI']
northeast = ['ME', 'VT', 'NH', 'MA', 'RI', 'CT', 'NJ', 'DE', 'MD', 'PA', "NY"]
southeast = ['WV', 'VA', 'TN', 'KY', 'KA', 'NC', 'SC', 'GA', 'FL', 'AL', 'MS', 'LA', 'AR']

d = {}

for s in west:
    d[s] = 'west'
for s in southwest:
    d[s] = 'southwest'
for s in midwest:
    d[s] = 'midwest'
for s in northeast:
    d[s] = 'northeast'
for s in southeast:
    d[s] = 'southeast'

def get_region(state):
    return d[state]

us_cases_by_region['state'] = us_cases_by_region['state'].apply(get_region)
#us_cases.loc[case_df['state'] == 'ME']
```

```python
#create scatterplot for cases in US

sns.lineplot(x="submission_date", y="tot_cases", data=us_cases, hue="state")
```

```python
#scatterplot of us cases by region 
sns.lineplot(x="submission_date", y="tot_cases", data=us_cases_by_region, hue="state").set(title='Number of COVID cases vs Date in the US', xlabel='Date', ylabel='Total Cases')
```
